```python
import pandas as pd
import re

def add_virtual_column(df: pd.DataFrame, role: str, new_column: str) -> pd.DataFrame:

    # Validation 1: Check if new_column name is valid (only letters and underscores)
    if not re.match(r'^[a-zA-Z_]+$', new_column):
        return pd.DataFrame()
    
    # Validation 2: Extract column names from the role expression
    # Find all potential column names (words that aren't operators)
    potential_columns = re.findall(r'[a-zA-Z_]\w*', role)
    
    # Check if all column names in the expression exist in the DataFrame
    for col in potential_columns:
        if col not in df.columns:
            return pd.DataFrame()
    
    # Validation 3: Check if role contains only valid operations
    # Remove column names and whitespace, then check if only valid operators remain
    role_check = role
    for col in df.columns:
        role_check = role_check.replace(col, '')
    role_check = role_check.replace(' ', '')
    
    # Check if remaining characters are only valid operators and parentheses
    valid_chars = set('+-*/()')
    remaining_chars = set(role_check)
    if not remaining_chars.issubset(valid_chars):
        return pd.DataFrame()
    
    try:
        # Create a copy of the DataFrame to avoid modifying the original
        result_df = df.copy()
        
        # Create a dictionary of column names to their values for eval
        eval_dict = {col: result_df[col] for col in result_df.columns}
        
        # Evaluate the expression and add the new column
        result_df[new_column] = eval(role, {"__builtins__": {}}, eval_dict)
        
        return result_df
    
    except Exception:
        # If any error occurs during evaluation, return empty DataFrame
        return pd.DataFrame()
```
